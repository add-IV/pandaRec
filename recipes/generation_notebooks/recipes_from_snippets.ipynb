```python
from pandaRec import Recipe, RecipeCollection
```

```python
df = pd.read_json('snippets.json').T
```

```python
df["name"] = df.index
```

```python
df.drop(columns=["scope", "prefix"], inplace=True)
```

```python
df.index = range(len(df))
```

```python
def create_recipe(id, row):
    code = "\n".join(row["body"])
    return {"id": id, "name": row["name"], "description": row["description"],"code": code, "keywords": ""}
```

```python
recipes = [create_recipe(idx, row) for idx, row in df.iterrows()]
recipes
```

```python
with open('recipes.json', 'w') as f:
    json.dump(recipes, f, indent=4)
```
