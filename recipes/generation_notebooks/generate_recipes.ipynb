```python
import inspect
import pandas as pd
from pandarec.recipe import Recipe
import yaml
```

```python
df = pd.DataFrame([[10, 20, 30], [100, 200, 300]],
                  columns=['foo', 'bar', 'baz'])

# modified from https://stackoverflow.com/a/34452
def get_methods(object):
  method_list = []
  for method in dir(object):
    if not callable(getattr(object, method)):
      continue
    method_name = str(method)
    if method_name.startswith('_'):
      continue
    if str(getattr(object, method).__doc__) == 'None':
      continue
    method_list.append([method_name, str(getattr(object, method).__doc__)])
  return method_list

methods = get_methods(df)

recipes = [Recipe(id, method[0], method[1].strip(),"", "") for id, method in enumerate(methods)]

for recipe in recipes:
  print(recipe)
```

```python
with open("./recipes2.yaml", 'w') as file:
    yaml.dump(recipes, file, sort_keys=False, )
```
