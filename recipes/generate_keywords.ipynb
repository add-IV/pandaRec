```python
import yake
import yaml
```

```python
with open("./examples/recipes2.yaml") as file:
    recipes = yaml.load(file, Loader=yaml.UnsafeLoader)
```

```python
descriptions = [recipe.description for recipe in recipes]
```

```python
kw_extractor = yake.KeywordExtractor()
keywords_per_recipe = [kw_extractor.extract_keywords(description) for description in descriptions]
```

```python
for keywords in keywords_per_recipe:
    print(keywords)
```

```python
i = 0
for keywords in keywords_per_recipe:
    for kw in keywords:
        i += 1
print(i)
```

```python
all_keywords = set([])
for keywords in keywords_per_recipe:
    for kw in keywords:
        all_keywords.add(kw[0])

print(len(all_keywords))
print(all_keywords)
```
