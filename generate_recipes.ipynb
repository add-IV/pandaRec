```python
import inspect
import pandas as pd
from pandaRec.recipe import Recipe
import yaml
```

```python
df = pd.DataFrame([[10, 20, 30], [100, 200, 300]],
                  columns=['foo', 'bar', 'baz'])

# modified from https://stackoverflow.com/a/34452
def get_methods(object):
  methodList = [str(method_name) for method_name in dir(object) if callable(getattr(object, method_name))]
  methodList = [method for method in methodList if not method.startswith('_')]
  # method.append(desc if (desc := str(getattr(object, method[0]).__doc__)) != 'None' else '')
  methodList = [[method, desc] if (desc := str(getattr(object, method).__doc__)) != 'None' else [method, ''] for method in methodList]
  return methodList

methods = get_methods(df['foo'])

recipes = [Recipe(id, method[0], method[1].strip(),"", "") for id, method in enumerate(methods)]

for recipe in recipes:
  print(recipe)
```

```python
with open("./recipes2.yaml", 'w') as file:
    yaml.dump(recipes, file, sort_keys=False, )
```
