```python
from datasets import load_from_disk
import pandas as pd
import matplotlib.pyplot as plt
```

```python
ds = load_from_disk('posts/combined/')
```

```python
occurences = {}

for data in ds:
    if data['Tags'] is None:
        continue
    for tag in data['Tags']:
        if tag in occurences:
            occurences[tag] += 1
        else:
            occurences[tag] = 1
```

```python
df = pd.DataFrame.from_dict(occurences, orient='index', columns=['Occurences'])
```

```python
df.sort_values(by=['Occurences'], ascending=False, inplace=True)
```

```python
df.to_csv('posts/tags.csv')
```

```python
plt.bar(df.index[:30], df['Occurences'][:30])
plt.xticks(rotation=90)
plt.show()
```
